## NASA Mars News

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://redplanetscience.com'
```

```python
browser.visit(url)
tm.sleep(3)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all('div', class_="slide")
results
```

```python
titles = soup.find_all('div', class_='content_title')
```

```python
titles
```

```python
latest_title = titles[0].text.strip()
```

```python
print(f'''The Latest Title on NASA Mars News Site:
{latest_title}''')
```

```python
para = soup.find_all('div', class_='article_teaser_body')
```

```python
para
```

```python
latest_para = para[0].text.strip()
```

```python
print(f'''The Latest paragraph on NASA Mars News Site:
{latest_para}''')
```

```python
browser.quit
```
